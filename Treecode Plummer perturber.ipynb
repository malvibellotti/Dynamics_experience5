# Let's now add a perturber

We want to initialize a perturber in orbit in a Plummer sphere with two different masses: $M_1 = 0.03M_{\text{tot}}$ and $M_2 = 0.01M_{\text{tot}}$ to study the different behaviours.

To initialize the position and the velocity of the perturber we need to make use of the equations for the conservation of energy and angular momentum on an unperturbed elliptical orbit with apoastron $R_{\text{apo}} = \frac{1}{3} R_{\text{peri}}$.

In particular, the equations that must be satisfied are:

$$ E(R_\text{apo};M(R_\text{apo})) = E(R_\text{peri};M(R_\text{peri}))   $$

$$ L(R_\text{apo}) = L(R_\text{peri})  $$

That can be written as:

$$ \frac{1}{2} \mu v_\text{apo}^2 - \frac{GM(R_\text{apo})\mu}{\sqrt{b^2+R_{\text{apo}}^2}} = \frac{1}{2} \mu v_\text{peri}^2 - \frac{GM(R_\text{peri})\mu}{\sqrt{b^2+R_{\text{peri}}^2}} $$

$$ \mu v_{\text{apo}} R_{\text{apo}} = \mu v_{\text{peri}} R_{\text{peri}} $$

Since $M_\text{Plummer} >> M_\text{perturber}$, we can approximate: $\mu \simeq M_\text{perturber}$ and $M(R) =  M_\text{Plummer}(R)$

Solving this system of equations we can get: $\vec{x_0} = (R_{\text{apo}},0,0)$ $\vec{v_0} = (0,v_{\text{apo}},0)$

Where $R_{\text{apo}}=6 \text{IU} = 6\text{pc}$ and $v_\text{apo} = \frac{1}{2}\Bigl(\frac{GM(R_\text{peri})}{\sqrt{b^2+R_\text{peri}^2}} - \frac{GM(R_\text{apo})}{\sqrt{b^2+9R_\text{peri}^2}}\Bigr)^{1/2}$

```python
def w_onfile(filename, num_points, masses, dimension, x, y, z):
    with open(filename, 'w') as file:
        file.write(f'  {num_points} \n  {dimension} \n  {0}  \n')
        #perturber mass
        file.write(f'  {(perturber_mass/M_iu).value}  \n')
        #mass of the others
        for _ in range(num_points-1):
            file.write(f'  {(masses/M_iu).value}  \n')
        #perturber's position
        file.write(f'  {x_pert}  {y_pert}  {z_pert}  \n')
        #position of the others
        for i in range(num_points-1):
            file.write(f'  {x[i]}  {y[i]}  {z[i]}  \n')
        #perturber's velocity
        #file.write(f'  {v_x_pert[0]}  {v_y_pert[0]}  {v_z_pert[0]} \n')
        file.write(f'  {v_x_pert}  {v_y_pert}  {v_z_pert} \n')
        #velocities of the others
        for i in range(num_points-1):
            #file.write(f'  {(vx[i]/v_iu).value}  {(vy[i]/v_iu).value}  {(vz[i]/v_iu).value}  \n')
            file.write(f'  {v_x[i]}  {v_y[i]}  {v_z[i]} \n')
```
