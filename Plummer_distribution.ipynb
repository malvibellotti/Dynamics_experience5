```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000 #number of particles
a = 5 #sphere radius
M = 1000 #sphere total mass
n_dim = 3 #number of dimensions
inputname = 'P_input3.txt'
outputname = 'P_output3.txt'
dyn_friction = True
```

We want to generate N particles distributed as a Plummer sphere. In order to do this, we use the inverse probability theorem: we generate points from a uniform distribution and pass them through the inverse formula of the original probability density.
We will do this in polar coordinates and then switch to cartesian ones to write the imput file.

### Radius distribution
Since we are talking about a Plummer distribution, the radius will be:

$$ r = \frac{a}{\sqrt{(P^{-2/3} - 1)}} $$

```python
P_r = np.random.uniform(0, 1, N)

radius = a / np.sqrt(P_r**(-2.0 / 3.0) - 1)
radius_s = radius.copy()
radius_s.sort()

fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot(121)
ax.plot(radius_s, np.linspace(0,1,N))
ax.set(xlabel='p(r)', ylabel='radius r', title='radius cfd') 

ax = fig.add_subplot(122)
ax.hist(radius, int(np.sqrt(N)), histtype = 'step', density = True)
ax.set(ylabel='p(r)', xlabel='radius r', title='radius histogram') 

plt.show()
```

### Phi distribution
$\phi$ is distributed uniformly and normalised with $2\pi$.

```python
phi = np.random.uniform(0, 2*np.pi, N)
phi_s = phi.copy()
phi_s.sort()

fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot(121)
ax.plot(phi_s, np.linspace(0,1,N))
ax.set(xlabel='p($\phi$)', ylabel='angle $\phi$', title='$\phi$ cfd') 

ax = fig.add_subplot(122)
ax.hist(phi, int(np.sqrt(N)), histtype = 'step')
ax.set(ylabel='p($\phi$)', xlabel='angle $\phi$', title='$\phi$ histogram')

plt.show()
```

### Theta distribution
Finally, $\theta$ is distributed as:
$$p(\theta) = sin(\theta)$$
and its inverse probability distribution is:
$$\theta = arccos(1 - 2P)$$

```python
P_theta = np.random.uniform(0, 1, N)
theta = np.arccos(1 - 2 * P_theta)
theta_s = theta.copy()
theta_s.sort()

fig = plt.figure(figsize = (15, 5))
ax = fig.add_subplot(121)
ax.plot(theta_s, np.linspace(0,1,N))
ax.set(xlabel='p(theta)', ylabel='angle theta', title='theta cfd') 

ax = fig.add_subplot(122)
ax.hist(theta, int(np.sqrt(N)), histtype = 'step')
ax.set(ylabel='p(theta)', xlabel='angle theta', title='theta') 

plt.show()
```

### Velocity distribution
In the Plummer sphere particles aren't at rest, but rather their velocities are distributed isotropically in space and with a certain module distribution. We will use rejection sampling to find the magnitudes.

```python
#generate the velocities

x = np.linspace(0, 1, N)
f1=(x**2)*(1-x**2)**(7/2)*512/(7*np.pi)   #512/7*pi is the normalitation?

def func (x):
    y=(x**2)*(1-x**2)**(7/2)*512/(7*np.pi)
    return y

x_goodpoints=[]  # in the notes is called q
y_goodpoints=[]

while len(x_goodpoints)<N:
    x1=np.random.uniform(0, 1)
    y1=np.random.uniform(0, max(f1))

    y2=func(x1)
    
    if y1 < y2:
        x_goodpoints.append(x1)
        y_goodpoints.append(y1)
    
plt.scatter(x_goodpoints,y_goodpoints)

plt.plot(x,f1, color='red')
plt.xlabel('q')
plt.ylabel('f(q)')
plt.show()

#converting q into velocities
pot=N/np.sqrt(a**2+radius**2)
v_module = x_goodpoints*np.sqrt(2*pot) #magnitude of the velocities

#sample of phi
v_phi=np.random.uniform(0, 2*np.pi, N)

#sample of theta
P_v_theta=np.random.uniform(0, 1, N)
v_theta = np.arccos(1 - 2 * P_theta)
```

### Creating the nbody input
We finally need to convert our values into carthesian coordinates and finally print them into a .txt file.

```python
def polar_to_carth(radius, theta, phi):
    x = radius * np.sin(theta) * np.cos(phi)
    y = radius * np.sin(theta) * np.sin(phi)
    z = radius * np.cos(theta)
    return x, y, z
```

```python
x, y, z = polar_to_carth(radius, theta, phi)
vx, vy, vz = polar_to_carth(v_module, v_theta, v_phi)

fig = plt.figure()

#points
ax = fig.add_subplot(projection='3d')
ax.scatter(0, 0, 0)
ax.scatter(x, y, z, s=1)

ax.set(xlabel='x', ylabel='y', zlabel='z', title=str(N) + ' points from a Plummer distr.', xlim = (-150, 150), ylim = (-150, 150), zlim = (-150, 150))

plt.show()

print(x.max(), y.max(), z.max())
```

### Adding a perturber

```python
m_perturber = M/N * 100
if M/N < m_perturber < M: print('mass condition satisfied')
m_x, m_y, m_z = 4, 4, 4
m_vx, m_vy, m_vz = -np.mean(vx) - np.std(vx), 0, 0
print(m_vx, m_vy, m_vz)

fig = plt.figure()

#points
ax = fig.add_subplot(projection='3d')
ax.scatter(m_x, m_y, m_z, s = 10, color = 'red')
ax.scatter(x, y, z, s=1, color = 'cornflowerblue')

ax.set(xlabel='x', ylabel='y', zlabel='z', title='Plummer distr. with perturber', xlim = (-20, 20), ylim = (-20, 20), zlim = (-20, 20))

plt.show()
```

### Creating the .txt file
We will need the number of particles (N), the number of dimentions (n_dim), the starting time (fixed to 0), the mass of the particles (all equal and chosen by the user depending on total mass), the points we have generated and finally the velocities.

```python
inputf = open(inputname, 'w')

particles_masses = ''
particles_positions = ''
particles_velocities = ''

if dyn_friction:
    particles_masses = particles_masses + str(m_perturber) + '\n'
    particles_positions = particles_positions + str(m_x) + ' ' + str(m_y) + ' ' + str(m_z) + '\n'
    particles_velocities = particles_velocities + str(m_vx) + ' ' + str(m_vy) + ' ' + str(m_vz) + '\n'
    
    for i in range(N):
        particles_masses = particles_masses + str(M/N) + '\n'
        particles_positions = particles_positions + str(x[i]) + ' ' + str(y[i]) + ' ' + str(z[i]) + '\n'
        particles_velocities = particles_velocities + str(vx[i]) + ' ' + str(vy[i]) + ' ' + str(vz[i]) + '\n'
    
    lines = [str(N+1) + '\n' + str(n_dim) + '\n0\n' + particles_masses + particles_positions + particles_velocities]
else:
    for i in range(N):
        particles_masses = particles_masses + str(M/N) + '\n'
        particles_positions = particles_positions + str(x[i]) + ' ' + str(y[i]) + ' ' + str(z[i]) + '\n'
        particles_velocities = particles_velocities + str(vx[i]) + ' ' + str(vy[i]) + ' ' + str(vz[i]) + '\n'
    
    lines = [str(N) + '\n' + str(n_dim) + '\n0\n' + particles_masses + particles_positions + particles_velocities]

inputf.writelines(lines)

inputf.close()
```

### Command line parameters
To put all together, we can now compute useful parameters to run the nbody code. We recall that the command line is as follows:

treecode IN = input_name OUT = output_name dTIME = h EPS = softening_value THETA = accuracy_parameter TSTOP = max_time DTOUT = output_frequency

```python
dtime = 0.002 #timestep
eps = 0.0767 #softening value
theta_p = 0.05 #accuracy parameter
tstop = 6 #max time
dtout = 0.1 #output frequency
n_repetitions = 1 #repetition of the system (start to collapse(1) and back(2))
```

```python
sphere_V = 4 * np.power(a, 3) / (3*np.pi)
sphere_density = (M)/sphere_V

dyn_time = np.sqrt((3 * np.pi)/(16 *sphere_density))
collapse_time = dyn_time * np.power(2, -1/2)
print(collapse_time)

commandf = open('command_line.txt', 'a')
commandf.write('\n./treecode in=' + inputname + ' out=' + outputname + ' dtime=' + str(dtime) + ' eps=' + str(eps) + ' theta=' + str(theta_p) + ' tstop=' + str(tstop) + ' dtout=' + str(dtout))
commandf.close()
```

### Physical values of useful quantities
We convert all the relevant quantities in physical values, as up to now we have expressed them in internal units.

```python
from astropy import constants as const
from astropy import units as u

G = const.G.cgs
kpc = const.kpc.cgs
M_sun = const.M_sun.cgs
```

```python
print('Total sphere mass = ', M * M_sun, '\nElement mass = ', M*M_sun/N)

sphere_V = 4 * np.power(a * kpc, 3) / (3*np.pi)
print('\nSphere volume = ', sphere_V)

sphere_density = (M * M_sun)/sphere_V
print('\nSphere density = ', sphere_density)

dyn_time = np.sqrt((3 * np.pi)/(16 * G * sphere_density))
print('\nDynamical time = ', dyn_time)

collapse_time = dyn_time * np.power(2, -1/2)

print('\nCollapse time = ', collapse_time)
```
